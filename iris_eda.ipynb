# Iris Dataset Exploratory Data Analysis

This notebook performs exploratory data analysis on the `data/Iris.csv` dataset. It covers data structure, summary statistics, visualizations, and high-level conclusions for the three Iris flower species.

## 1. Setup and Data Loading

Load the dataset, inspect the first rows, and verify the schema.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
iris_path = 'data/Iris.csv'
iris = pd.read_csv(iris_path)

# Basic overview
iris.head()
```

### Data structure and basic checks

```python
print('Shape:', iris.shape)
print('Columns:', iris.columns.tolist())
print('Species values:', iris['Species'].unique())
print('Missing values by column:
', iris.isna().sum())

iris.info()
```

## 2. Summary Statistics

Explore the numeric distribution of sepal and petal measurements.

```python
iris.describe()
```

### Per-species summary
Compare the central tendencies and ranges for each Iris species.

```python
iris.groupby('Species').agg(['mean', 'median', 'std']).T
```

## 3. Correlation and Relationships

Examine relationships among numeric features and identify the strongest predictors.

```python
corr_matrix = iris.drop(columns=['Id']).corr()
corr_matrix
```

### Pairwise relationships
Visualize the distribution and pairwise relationships across features.

```python
sns.set(style='whitegrid', palette='colorblind')
pairplot_fig = sns.pairplot(iris.drop(columns=['Id']), hue='Species', markers=['o', 's', 'D'])
pairplot_fig.fig.suptitle('Iris Pairplot by Species', y=1.02)
plt.show()
```

## 4. Feature Distributions per Species

Compare each measurement across species using boxplots and violin plots.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(data=iris, x='Species', y=feature, ax=ax)
    ax.set_title(f'{feature} by Species')
    ax.set_xlabel('')
plt.tight_layout()
plt.show()
```

## 5. Additional Visualizations

Use violin plots to compare the density of feature values by species.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
for ax, feature in zip(axes.flatten(), features):
    sns.violinplot(data=iris, x='Species', y=feature, inner='quartile', ax=ax)
    ax.set_title(f'{feature} Distribution by Species')
    ax.set_xlabel('')
plt.tight_layout()
plt.show()
```

## 6. Conclusions

Summarize the most important findings from the Iris dataset analysis.

- The dataset contains 150 observations and 5 numeric features plus species labels.
- There are three species: `Iris-setosa`, `Iris-versicolor`, and `Iris-virginica`.
- `PetalLengthCm` and `PetalWidthCm` differ most clearly across species and are strong separators.
- `SepalWidthCm` is less distinctive between species than petal measurements.
- Pearson correlation indicates strong relationships between petal length and petal width, while sepal measurements show weaker correlation.
- Visualizations confirm that `Iris-setosa` is separable from the other two classes, whereas `Iris-versicolor` and `Iris-virginica` are more similar but still distinguishable by petal dimensions.
